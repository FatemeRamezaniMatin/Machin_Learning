```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage
from scipy.spatial import distance_matrix
from sklearn import datasets, manifold
from scipy.cluster import hierarchy
from sklearn.cluster import AgglomerativeClustering
from sklearn.datasets import make_blobs
```

## Random Data

```python
X,y = make_blobs(n_samples = 50, centers= [[-1,2],[3,4],[10,5]], cluster_std = 0.9)
```

```python
plt.scatter(X[:,0],X[:,1], marker='o')
```

## Agglomerative Clustering

```python
agglomer = AgglomerativeClustering(n_clusters = 4, linkage = 'average')
agglomer.fit(X,y)
```

```python
plt.figure(figsize=(6,4))
x_min , x_max = np.min(X, axis=0), np.max(X, axis=0)
X = (X- x_min) / (x_max - x_min)
for i in range(X.shape[0]):
    plt.text(X[i, 0], X[i, 1], str(y[i])),
    color=plt.cm.nipy_spectral(agglomer.labels_[i] / 10.),
    fontdict = {'weight':'bold', 'size': 9}
plt.xticks([])
plt.yticks([])
plt.scatter(X[:,0], X[:,1], marker= '.')
plt.show()
```

## Dendrogram Associated for the Agglomerative Hierarchical Clustering

```python
Z = hierarchy.linkage(dist_matrix, 'ward')
```

```python
dendro = hierarchy.dendrogram(Z)
```

## Read Data

```python
import pandas as pd
file_path = "cars_clus.csv"
cars_df = pd.read_csv(file_path)
cars_df.head()
```

```python
print('shape of dataset before cleaning:', cars_df.shape)
```

## Data Cleaning

```python
cars_df[['sales', 'resale',	'type' , 'price', 'engine_s', 'horsepow', 'wheelbas', 'width' ,
         'length', 'curb_wgt', 'fuel_cap', 'mpg', 'lnsales']]= cars_df[['sales', 'resale',	'type' , 'price', 'engine_s', 'horsepow', 'wheelbas', 
         'width' ,'length', 'curb_wgt', 'fuel_cap', 'mpg', 'lnsales']].apply(pd.to_numeric, errors='coerce')
cars_df = cars_df.dropna()
cars_df = cars_df.reset_index(drop=True)
cars_df['kml'] = 236.25/cars_df['mpg']
print('shape of dataset after cleaning:', cars_df.shape)
cars_df.head()
```

## Feature Selection

```python
featureset = cars_df[['engine_s', 'horsepow', 'wheelbas', 'width', 'length', 'curb_wgt', 'fuel_cap', 'mpg', 'kml']]
```

## Normalization

```python
from sklearn.preprocessing import MinMaxScaler
X = featureset.values
min_max = MinMaxScaler()
feature_mtx = min_max.fit_transform(X)
feature_mtx [0:5]
```

## Clustering

```python
from scipy.spatial.distance import cdist
D = cdist(feature_mtx, feature_mtx, metric='euclidean')
D
```

```python
import pylab
import scipy.cluster.hierarchy
Z = hierarchy.linkage(D, 'complete')
```

```python
from scipy.cluster.hierarchy import fcluster
max_d = 3
clusters_d = fcluster(Z, max_d, criterion='distance')
clusters_d
```

```python
k = 5
clusters_k = fcluster(Z, k, criterion='maxclust')
clusters_k
```

```python
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.scatter(feature_mtx[:,0], feature_mtx[:,1], c=clusters_d, cmap='tab10')
plt.title("Distance")

plt.subplot(1, 2, 2)
plt.scatter(feature_mtx[:,0], feature_mtx[:,1], c=clusters_k, cmap='tab10')
plt.title("Maxclust")
plt.show()
```

```python
fig = pylab.figure(figsize=(18,50))
def llf(id):
    return '[%s %s %s]' % (cars_df['manufact'][id], cars_df['model'][id], int(float(cars_df['type'][id])))

dendro = hierarchy.dendrogram(Z, leaf_label_func=llf, leaf_rotation=0, orientation='right', leaf_font_size=12)
```

## Clustring using scikit_learn 

```python
from sklearn.metrics.pairwise import euclidean_distances
dist_metrics = euclidean_distances(feature_mtx, feature_mtx)
print(dist_metrics)
```

```python
Z_using_dist_metrics = hierarchy.linkage(dist_metrics, 'complete')
```

```python
fig = pylab.figure(figsize=(18,50))
def llf(id):
    return '[%s %s %s]' % (cars_df['manufact'][id], cars_df['model'][id], int(float(cars_df['type'][id])))

dendro = hierarchy.dendrogram(Z, leaf_label_func=llf, leaf_rotation=0, orientation='right', leaf_font_size=12)
```

```python
agglom = AgglomerativeClustering(n_clusters=6, linkage='complete')
agglom.fit(dist_metrics)

agglom.labels_
```

```python
cars_df['clusters'] = agglom.labels_
cars_df.head()
```

```python
import seaborn as sns

plt.figure(figsize=(14, 14))

sns.scatterplot(
    data=cars_df,
    x='horsepow',
    y='kml',
    hue='clusters',
    size='price',
    sizes=(20, 800),
    palette='hsv',
    alpha=0.6
)

for _, row in cars_df.iterrows():
    plt.text(row.horsepow, row.kml, str(row['model']), rotation=25, fontsize=9)

plt.title("Car Clusters")
plt.xlabel("Horsepower")
plt.ylabel("KML")
plt.legend(title='Cluster')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
cars_df.groupby(['clusters', 'type'])['clusters'].count()
```

```python
agg_cars = cars_df.groupby(['clusters','type'])[['horsepow','engine_s','kml','price']].mean()
agg_cars
```

```python
plt.figure(figsize=(16, 10))

sns.scatterplot(
    data=agg_cars.reset_index(),  
    x='horsepow',
    y='kml',
    hue='clusters',
    size='price',      
    sizes=(50, 1000),   
    palette='tab10',
    alpha=0.6
)

for _, row in agg_cars.reset_index().iterrows():
    plt.text(
        row.horsepow + 5,
        row.kml,
        f"type={row['type']}, price={int(row['price'])}k",
        fontsize=9,
        rotation=25
    )

plt.title("Clusters")
plt.xlabel("horsepow")
plt.ylabel("kml")
plt.legend(title="Cluster")
plt.grid(True)
plt.tight_layout()
plt.show()
```

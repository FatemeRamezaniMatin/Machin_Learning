```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

## Understanding Data

```python
file_path = "customer.csv"
cust_df = pd.read_csv(file_path)
cust_df.head()
```

```python
cust_df.shape
```

```python
cust_df.info()
```

```python
cust_df.isnull().sum()
```

```python
cust_df.describe()
```

```python
from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()
cust_df['Gender'] = label.fit_transform(cust_df['Gender']) 
cust_df.head() # Male=1, Female=0
```

```python
cust_df.isnull()
```

## Preprocessing

```python
df = cust_df.drop('CustomerID', axis=1)
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
X = df.values
cust_dataset = StandardScaler().fit_transform(X)
cust_dataset
```

## Elbow Method

```python
from sklearn.cluster import KMeans
inertia=[]
K = range(1,11)
for k in K:
    k_means = KMeans(n_clusters=k, random_state=42)
    k_means.fit(X)
    inertia.append (k_means.inertia_)

plt.plot(list(K), inertia, 'bo-')
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Inertia")
plt.title('Elbow Method')
plt.show()
```

## Silhouette Score

```python
from sklearn.metrics import silhouette_score
k = range(2,11)
for K in k:
    k_means = KMeans(n_clusters = K, random_state=42)
    cluster_label = k_means.fit_predict(X)
    silhouette_ave = silhouette_score(X, cluster_label)
    print("for n_clusters",K ,':',silhouette_ave)
```

## Modeling __ KMeans

```python
K_Means = KMeans(init="k-means++", n_clusters=7, random_state=42, n_init=12)
K_Means.fit(cust_dataset)
labels = K_Means.labels_
labels
```

```python
df['Cust_km'] = labels
df.head()
```

```python
df.groupby('Cust_km').mean()
```

## Visualization

```python
plt.figure(figsize=(10,6))
df.groupby('Cust_km').mean(numeric_only=True).plot(kind='bar')
plt.xticks(rotation=0)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='Annual Income (k$)', y='Spending Score (1-100)', hue='Cust_km', palette='Set2', s=100)
plt.show()
```

## Modeling __ Hierarchical

```python
from sklearn.cluster import AgglomerativeClustering
agglom = AgglomerativeClustering(n_clusters=7, linkage='average')
agglom.fit(cust_dataset)

agglom.labels_
```

```python
df['Cust_Hie'] = agglom.labels_
df.head()
```

```python
df.groupby('Cust_Hie').mean()
```

```python
from scipy.cluster.hierarchy import cophenet, linkage, dendrogram
```

```python
from scipy.spatial.distance import pdist
```

```python
Z = linkage(cust_dataset, method='average', metric='euclidean')
```

```python
c, coph_dists = cophenet(Z, pdist(cust_dataset))
c
```

## Visualization

```python
plt.figure(figsize=(10,50))
dendrogram(Z, leaf_rotation=0, leaf_font_size=10, orientation='right')
plt.title('Dendrogram')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='Annual Income (k$)', y='Spending Score (1-100)', hue='Cust_Hie', palette='Set1', s=100)
plt.show()
```

## Modeling__DBSCAN

```python
from sklearn.neighbors import NearestNeighbors
neigh = NearestNeighbors(n_neighbors=5)  # min_samples = 5
nbrs = neigh.fit(cust_dataset)
distances, indices = nbrs.kneighbors(cust_dataset)
distances = np.sort(distances[:, 3])
plt.figure(figsize=(10, 5))
plt.plot(distances)
plt.grid(True)
plt.show()
```

```python
from sklearn.manifold import TSNE
tsne = TSNE(n_components=3, random_state=42)
cust_dataset_tsne = tsne.fit_transform(cust_dataset)
```

```python
from sklearn.cluster import DBSCAN
dbscan = DBSCAN(eps=0.78, min_samples=5)
dbscan.fit(cust_dataset_tsne)
labels = dbscan.labels_
labels
```

## Distinguish Outlier

```python
core_sample_mask = np.zeros_like(labels, dtype=bool)
core_sample_mask[dbscan.core_sample_indices_] = True
core_sample_mask
```

```python
n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
n_clusters
```

```python
unique_labels = set(labels)
unique_labels
```

## Visualization

```python
color = plt.cm.Spectral(np.linspace(0, 1, len(unique_labels)))
for K, col in zip(unique_labels, color):
    if K == -1: 
        col ='k'

    class_member_mask = (labels == K)

    xy = cust_dataset[class_member_mask & core_sample_mask]
    plt.scatter(xy[:,1], xy[:,2], s=50, c=[col], marker='o', alpha=0.7)

    xy = cust_dataset[class_member_mask & ~core_sample_mask]
    plt.scatter(xy[:,1], xy[:,2], s=50, c=[col], marker='o', alpha=0.7)
```

```python
print("Silhouette Score (KMeans):", silhouette_score(cust_dataset, K_Means.labels_))
print("Silhouette Score (Hierarchical):", silhouette_score(cust_dataset, agglom.labels_))
print("Silhouette Score (DBSCAN):", silhouette_score(cust_dataset, dbscan.labels_))
```

```python
from sklearn.metrics import davies_bouldin_score

print("Davies-Bouldin Index (KMeans):", davies_bouldin_score(X, K_Means.labels_))
print("Davies-Bouldin Index (Hierarchical):", davies_bouldin_score(X, agglom.labels_))
print("Davies-Bouldin Index (DBSCAN):", davies_bouldin_score(X, dbscan.labels_))
```

```python
fig, axs = plt.subplots(1, 3, figsize=(18, 5))

axs[0].scatter(cust_dataset[:, 1], cust_dataset[:, 2], c=K_Means.labels_, cmap='rainbow')
axs[0].set_title('KMeans')

axs[1].scatter(cust_dataset[:, 1], cust_dataset[:, 2], c=agglom.labels_, cmap='rainbow')
axs[1].set_title('Hierarchical')

axs[2].scatter(cust_dataset[:, 1], cust_dataset[:, 2], c=dbscan.labels_, cmap='rainbow')
axs[2].set_title('DBSCAN')

plt.show()
```
